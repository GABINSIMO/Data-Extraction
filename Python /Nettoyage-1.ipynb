*Nettoyage du fichier Rendez-vous (appointments)*

```python
import pandas as pd

appointments = "/Users/simogabin/Documents/Projets/Extraction des données /Dataset/archive/appointments.csv"

df = pd.read_csv(appointments)

# Vérifier les doublons
df = df.drop_duplicates()

# Vérifier les valeurs manquantes
print(df.isnull().sum())

# Convertir les colonnes en type correct
df["appointment_date"] = pd.to_datetime(df["appointment_date"])
df["appointment_time"] = pd.to_datetime(df["appointment_time"], format="%H:%M:%S").dt.time

# Normaliser le texte des colonnes
df["status"] = df["status"].str.strip().str.capitalize()
df["reason_for_visit"] = df["reason_for_visit"].str.strip().str.capitalize()

print(df.head())
```

```python
# sauvegarder le df nettoyer
df.to_csv("/Users/simogabin/Documents/Projets/Extraction des données /Dataset/Dataset_clean/RDV.csv", index=False)
```

*Nettoyage du fichier Facture (billings)*

```python
import pandas as pd

billings = '/Users/simogabin/Documents/Projets/Extraction des données /Dataset/archive/billing.csv'

df = pd.read_csv(billings)

# Vérifier les doublons
df = df.drop_duplicates()

# Vérifier les valeurs manquantes
print(df.isnull().sum())

# Convertir les colonnes en type correct
df["bill_date"] = pd.to_datetime(df["bill_date"])
df["amount"] = df["amount"].map("{:,.2f}".format)


# Normaliser le texte des colonnes
df["payment_status"] = df["payment_status"].str.strip().str.capitalize()
df["payment_method"] = df["payment_method"].str.strip().str.capitalize()

print(df.head())
```
